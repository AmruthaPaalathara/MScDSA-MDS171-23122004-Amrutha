```python
import random
item=["Potato","Tomato","Raddish","Onion","Garlic","Cucumber","Chilli","Yam","Carrot","Spinach"]  # list of item
quantity=["1",'500','250','1','100','50','100','500','300','150']   # quantity of the item
price=['150','100','130','200','130','50','90','45','65','300']     # price of the item
file=open("23122004.csv","w+")
for i in range(100):
    r1=random.randint(0,9)
    r2=random.randint(0,9)
    r3=random.randint(0,9)
    Quantity=int(quantity[r2])
    Price=int(price[r3])
    total=Quantity*Price
    grand_total=str(total)
    file.write(item[r1]+','+quantity[r2]+','+price[r3]+'\t, Total ='+grand_total+"\n")
file.close()
               
```

```python
with open("23122004.csv","r+") as file:
    potatotot=0
    tomatotot=0
    raddishtot=0
    oniontot=0
    garlictot=0
    cucumbertot=0
    chillitot=0
    yamtot=0
    carrotot=0
    spinachtot=0

    qpotato=0
    qtomato=0
    qraddish=0
    qonion=0
    qgarlic=0
    qcucumber=0
    qchilli=0
    qyam=0
    qcarrot=0
    qspinach=0
    dict1={"Potato":[],"Tomato":[],"Raddish":[],"Onion":[],"Garlic":[],"Cucumber":[],"Chilli":[],"Yam":[],"Carrot":[],"Spinach":[]}
    for line in file:
        lines=line.strip().split("=")
        qline=line.strip().split(",")
        
        
        if "Potato" in line:
            qp=int(qline[1])
            p=int(lines[1])
            potatotot+=p
            qpotato+=qp
            dict1["Potato"].append(int(qline[2].strip()))
        elif "Tomato" in line:
            qt=int(qline[1])
            t=int(lines[1])
            tomatotot+=t
            qtomato+=qt
            dict1["Tomato"].append(int(qline[2].strip()))
        elif "Raddish" in line:
            qr=int(qline[1])
            r=int(lines[1])
            raddishtot+=r
            qraddish+=qr
            dict1["Raddish"].append(int(qline[2].strip()))
        elif "Onion" in line:
            qo=int(qline[1])
            o=int(lines[1])
            oniontot+=o
            qonion+=qo
            dict1["Onion"].append(int(qline[2].strip()))
        elif "Garlic" in line:
            qg=int(qline[1])
            g=int(lines[1])
            garlictot+=g
            qgarlic+=qg
            dict1["Garlic"].append(int(qline[2].strip()))
        elif "Cucumber" in line:
            qcu=int(qline[1])
            cu=int(lines[1])
            cucumbertot+=cu
            qcucumber+=qcu
            dict1["Cucumber"].append(int(qline[2].strip()))
        elif "Chilli" in line:
            qc=int(qline[1])
            c=int(lines[1])
            chillitot+=c
            qchilli+=qc
            dict1["Chilli"].append(int(qline[2].strip()))
        elif "Yam" in line:
            qy=int(qline[1])
            y=int(lines[1])
            yamtot+=y
            qyam+=qy
            dict1["Yam"].append(int(qline[2].strip()))
        elif "Carrot" in line:
            qc=int(qline[1])
            c=int(lines[1])
            carrotot+=c
            qcarrot+=qc
            dict1["Carrot"].append(int(qline[2].strip()))
        elif "Spinach" in line:
            qs=int(qline[1])
            s=int(lines[1])
            spinachtot+=s
            qspinach+=qs
            dict1["Spinach"].append(int(qline[2].strip()))
        else:
            pass
     
print(dict1)  
```
