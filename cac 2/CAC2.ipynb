
**EDA BIVARIATE ANALYSIS**
 
 In bivariate analysis we shall try to find whether any two factors from the given dataset are related in some way.

 Since the factors are categorical in nature, the test we are using to check for the correlation between 2 factors is #Chi-square test.

There will be two hypotheses:

H0: The factors are correlated
vs.
H1: The factors are uncorrelated

After applying Chi-square test, we calculate the p value. If the p-value < 0.05, we reject H0.


 

```python

CrosstabResult=pd.crosstab(index=cac['AVAILABILITY OF MEDICINES AT LAVASA'],columns=cac['RATE THE MEDICAL SERVICES AVAILABLE IN LAVASA .'])
print(CrosstabResult)

# importing the required function
from scipy.stats import chi2_contingency

# Performing Chi-sq test
ChiSqResult = chi2_contingency(CrosstabResult)
 
print('The P-Value of the ChiSq Test is:', ChiSqResult[1])


# Interpretation : Availablity of medicines and the rating of medical services is highly correlated because the p-value is quite less than 0.05. Also, after seeing previous charts of these two factors we can say that due to the poor availablity of the medicines in Lavasa affects the overall health services rating( which is mostly poor or average) 
```

```python

CrosstabResult=pd.crosstab(index=cac['DID YOU HAVE ANY HEALTH ISSUES AFTER COMING TO LAVASA ?'],columns=cac['RATE THE MEDICAL SERVICES AVAILABLE IN LAVASA .'])
print(CrosstabResult)

# importing the required function
from scipy.stats import chi2_contingency

# Performing Chi-sq test
ChiSqResult = chi2_contingency(CrosstabResult)
 
print('The p-Value of the ChiSq Test is:', ChiSqResult[1])

# Interpretation : The above 2 factors are not related since the p-value>0.05
```
